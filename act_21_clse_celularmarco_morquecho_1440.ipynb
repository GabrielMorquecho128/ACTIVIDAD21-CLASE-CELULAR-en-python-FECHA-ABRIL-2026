```python
class Celular:
    # Método constructor para inicializar atributos
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    # Método para mostrar información
    def mostrar_info(self):
        return f"{self.marca} {self.modelo} - ${self.precio}"

    # Método para encender el celular
    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} está encendido.")

# Instanciar el objeto
celular1 = Celular("Samsung", "Galaxy S21", 799.99)
celular2 = Celular("Xiaomi", "Redmi Note 10", 399.99)
# llamada a las funciones para cada objeto celular 1 y celular 2
print("celular1:")
print(celular1.mostrar_info())
celular1.encender()

print("\ncelular2")
print(celular2.mostrar_info())
celular2.encender()

print("programa terminado")
print("crear software marco morquecho NL= 39")
```
